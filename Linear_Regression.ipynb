```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_dataset.csv")
df.head()
```

```python
df.info()
```

```python
x = df[["YearsExperience"]]                   # feature or independent variable
y = df['Salary']                              # target or depenndent variable
```

```python
from sklearn.model_selection import train_test_split

Xtrain,Xtest,Ytrain,Ytest = train_test_split(x,y,train_size=0.8,random_state=42)
Xtrain
Xtest
```

# to understand the pattern
wetaher linear or non-linear data 

```python
plt.plot(x,y)
```


# model creation 

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

# model training

```python
model.fit(Xtrain,Ytrain)                                #y=mx+c
```

# model evaluating

```python
model.score(Xtest,Ytest)                                # it is used to find weather the tranied model has good or bad result
```

# predicting for actual data

```python
predict = model.predict(Xtest)  
```

```python
plt.plot(Xtest,predict)
plt.plot(x,y)
```

# future prediction

```python

new_data = pd.DataFrame({"YearsExperience":[50,100,110]})
model.predict(new_data)
```

```python
model.coef_
model.intercept_
```

```python
from sklearn.metrics import r2_score

r2_score(Ytest,predict)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error

ypred=model.predict(Xtest)

mean_squared_error(Ytest,ypred)                                
                       
```

```python
mean_absolute_error(Ytest,ypred)                     
```

1. mean_squared_error and mean_absolute_error, is one of the model evaluation methods
2. If the range of the target data is less we can use MAE or MSE
3. If the range of the target data is more we should use MAE
